This code fills missing values in the 'Arrival Delay in Minutes' column with 10 but only for rows where there is a 'Departure Delay in Minutes' greater than 0 and where the 'Arrival Delay in Minutes' is currently missing.

If a flight was delayed in departure and the arrival delay information is missing, it assumes the arrival delay was 10 minutes and fills in that value.

```python
#filling the missing values based on specifing the conditions
data.loc[(data['Departure Delay in Minutes'])&(data['Arrival Delay in Minutes'].isnull()),'Arrival Delay in Minutes']=10
data
```

The code identifies outliers in the 'Flight Distance' column of your dataset using the Interquartile Range (IQR) method.In essence, the code finds unusual 'Flight Distance' values that are significantly higher or lower than the typical range of distances in the data

```python
#Finding Outliers by using IQR
import numpy as np
#Calculate Q1 and Q3
Q1=data['Flight Distance'].quantile(0.25)
Q3=data['Flight Distance'].quantile(0.75)
#Calculate IQR
IQR=Q3-Q1
#Defining the upper bounds and lower bounds for outliers
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
#Filtering the data inorder to identify the outliers
IQR_outliers=data[(data['Flight Distance']<lower_bound)|(data['Flight Distance']>upper_bound)]
IQR_outliers
#Displaying those outliers
print("Outliers are detected by using IQR method")
print(IQR_outliers['Flight Distance'])

```
